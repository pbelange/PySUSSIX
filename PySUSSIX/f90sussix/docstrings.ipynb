```python

# Help from chatGPT
import re

def extract_docstring(filename, subroutine_or_function_name):
    with open(filename, 'r') as file:
        content = file.read()

    # Use a flexible regular expression to find the subroutine or function and its docstring
    pattern = re.compile(r'\s*(subroutine|function)\s+' + subroutine_or_function_name + r'\s*\([\s\S]*?end\s*(subroutine|function)', re.IGNORECASE)
    match = pattern.search(content)

    if match:
        docstring_pattern = re.compile(r'!={2,}([\s\S]*?)!={2,}', re.IGNORECASE)
        docstring_match = docstring_pattern.search(match.group())

        if docstring_match:
            docstring = docstring_match.group(1).strip()
            return docstring

    return None

# filename = '../src/sussixNoO.f'
# subroutine_or_function_name = 'tunenewt'


# if docstring:
#     print("Docstring for subroutine or function {}:\n{}".format(subroutine_or_function_name, docstring))
# else:
#     print("Subroutine or function {} not found or docstring not present.".format(subroutine_or_function_name))
```
